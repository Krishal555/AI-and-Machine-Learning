**• Conversion Options:**

2. Weight:
 Convert kilograms (kg) to pounds (lbs).
 Convert pounds (lbs) to kilograms (kg).
3. Volume:
 Convert liters (L) to gallons (gal).
 Convert gallons (gal) to liters (L).


```python
def convert_units(conversion_type, value):
    """
    Converts units based on selected conversion type.

    Parameters:
    conversion_type (str): Type of conversion
    value (float): Value to convert

    Returns:
    float: Converted value
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type")


try:
    print("Choose conversion:")
    print("1. meters to feet")
    print("2. feet to meters")
    print("3. kg to pounds")
    print("4. pounds to kg")
    print("5. liters to gallons")
    print("6. gallons to liters")

    choice = int(input("Enter choice: "))
    value = float(input("Enter value to convert: "))

    conversions = {
        1: "m_to_ft",
        2: "ft_to_m",
        3: "kg_to_lbs",
        4: "lbs_to_kg",
        5: "l_to_gal",
        6: "gal_to_l"
    }

    result = convert_units(conversions[choice], value)
    print("Converted value:", result)

except ValueError:
    print("Invalid input. Please enter numeric values.")
except KeyError:
    print("Invalid conversion choice.")
```

# 4.3 Exercise on Nested List:

**1. Flatten a Nested List:**
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.
**• Requirements:**
 Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.
 Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].


```python
def flatten(lst):
    """Flattens one level nested list."""
    result = []
    for i in lst:
        result.extend(i)
    return result


def access_nested_element(lst, indices):
    """Access element using list of indices."""
    element = lst
    for i in indices:
        element = element[i]
    return element


def sum_nested(lst):
    """Returns sum of all elements in nested list."""
    total = 0
    for i in lst:
        if isinstance(i, list):
            total += sum_nested(i)
        else:
            total += i
    return total


def remove_element(lst, elem):
    """Removes element from nested list."""
    result = []
    for i in lst:
        if isinstance(i, list):
            result.append(remove_element(i, elem))
        else:
            if i != elem:
                result.append(i)
    return result


def find_max(lst):
    """Find maximum element in nested list."""
    max_val = float('-inf')
    for i in lst:
        if isinstance(i, list):
            max_val = max(max_val, find_max(i))
        else:
            max_val = max(max_val, i)
    return max_val


def count_occurrences(lst, elem):
    """Counts occurrences in nested list."""
    count = 0
    for i in lst:
        if isinstance(i, list):
            count += count_occurrences(i, elem)
        else:
            if i == elem:
                count += 1
    return count


def deep_flatten(lst):
    """Flattens nested list of any depth."""
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)
    return result


def average_nested(lst):
    """Returns average of nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)
nested = [[1,2],[3,[4,5]],6]

print("Flatten:", flatten([[1,2],[3,4],[5]]))
```

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1 Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
I = np.dot(A,A_inv)

print("A * A^-1 =\n", I)

# 2 Prove AB ≠ BA
AB = np.dot(A,B)
BA = np.dot(B,A)

print("AB=\n", AB)
print("BA=\n", BA)

# 3 Prove (AB)^T = B^T A^T
left = (AB).T
right = np.dot(B.T, A.T)

print("(AB)^T=\n", left)
print("B^T A^T=\n", right)
```
